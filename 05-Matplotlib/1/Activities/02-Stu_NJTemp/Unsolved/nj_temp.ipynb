```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
avg_temp = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
months = np.arange(1, 13, 1)
months
```

```python
plt.plot(months, avg_temp)
plt.show()
```

```python
celsius_temp = [(x - 32) * 0.56 for x in avg_temp]
celsius_temp
```

```python
plt.plot(months, celsius_temp)
plt.show()
```

```python
plt.plot(months, avg_temp)
plt.plot(months, celsius_temp)

plt.show()
```

```python
temp_handle, = plt.plot(months, avg_temp, marker='+', color='blue', label='Fahrenheit')
celsius_handle, = plt.plot(months, celsius_temp, marker='s', color='red', label='Celsius')

plt.xlabel("Months")
plt.ylabel("Temperature")

plt.legend(loc='upper right')
```
