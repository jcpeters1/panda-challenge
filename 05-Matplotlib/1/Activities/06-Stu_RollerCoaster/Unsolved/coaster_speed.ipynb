```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.arange(0, 130, 10)
speed_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
speed_rail = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
plt.plot(time, speed_drop)
plt.plot(time, speed_rail)
plt.show()
```

```python
drop, = plt.plot(time, speed_drop, color='red', label='Danger Drop')
rail, = plt.plot(time, speed_rail, color='green', label='RailGun')
plt.xlabel('Coaster Runtime')
plt.ylabel('Speed (MPH)')
plt.title('Coaster Speed Over Time')
plt.legend(loc='upper right')
plt.grid()
plt.ylim(5,95)
plt.xlim(0,120)
```
